### Выдвижение гипотезы

По виду графика эмпирической функции распределения  и гистограмм выдвигаем двухальтернативную гипотезу о законе распределения случайной величины X:

H0 – величина X распределена по нормальному закону:
f(x) = f0(x) = (1/σ(2п)^0.5)e^(-((x-m)^2)/2σ^2);
F(x) = F0(x) = 0.5 + Ф((x-m)/σ);

H1 – величина X не распределена по нормальному закону
f(x) != f0(x)
F(x) != F0(x)

Определим оценки неизвестных параметров m  и σ гипотетического (нормального) закона распределения по формулам: m = Mx; σ = (Dx)^0.5

```python
e = 2.718
Q = Dx**0.5
```

Проверим гипотезу о нормальном законе по критерию Пирсона X^2. Вычислим значение критерия X^2 на основе равноинтервального статистического ряда:

X^2 = n * sum(((p_j - p_j*)^2)/p_j)

k = M-s-1

### Критерий Колмогорова

Проверим гипотезу при помощи критерия Колмогорова. Для этого построим  график гипотетической функции распределения  в одной системе координат с эмпирической функцией. В качестве опорных точек используем 10 значений A_j. По графику определим максимальное по модулю отклонение между функциями F*(x) и Fo(x):

```python
x = sorted_data
y = [i/len(data) for i in range(0, 100, 1)]
y2 = [i/len(data) for i in range(0, 100, 10)]
y2.append(y[-1])
x2 = copy.deepcopy(A_j)
x2.append(intervals[-1])

plt.figure(figsize=(12, 7))
plt.plot(x2, y2, 'r', alpha=0.7, label="Fo(X)",  lw=3)
plt.plot(x, y, 'o', alpha=0.7, label="Эмпирическая функция распределения F*(x)", lw=5, mec='b')
plt.legend()
plt.grid(True)
```

Z = max|F*(xi)-Fo(xi)|

```python
def get_dist(pointX,pointY,lineX1,lineY1,lineX2,lineY2):
    a=lineY2-lineY1
    b=lineX1-lineX2
    c=lineX2*lineY1-lineX1*lineY2
    dis=(math.fabs(a*pointX+b*pointY+c))/(math.pow(a*a+b*b,0.5))
    return dis
```

```python
dist = [round(get_dist(x,y,x2[0],y2[0],x2[-1],y2[-1]), 3) for x,y in zip(x,y)]
Z = max(dist)
Z
```

### Вычисление значения критерия Колмогорова

Lambda = n^0.5 * Z

```python
Lambda = round(len(data)**0.5 * Z, 3)
Lambda
```

Из таблицы Колмогорова по заданному уровню значимости выбираем критическое значение критерия: Lambda_y = 1.36

```python
Lambda_y = 1.36
print(Lambda <= Lambda_y)
if Lambda <= Lambda_y:
    print(f'''{Lambda} <= {Lambda_y} 
Так как условие критерия Колмогорова выполняется, 
гипотеза H¬0 о нормальном законе распределения принимается 
(нет оснований ее отклонить).''')
else:
    print(f'''{Lambda} > {Lambda_y} 
Так как условие критерия Колмогорова не выполняется,
гипотеза H0 о нормальном законе распределения не принимается.''')
```

#### Оценки математических ожиданий по каждой переменной

```python
# Мат. ожидание: Mx =1/n *(sum(Xi))
Mx2 = round((1/len(x2)*sum(x2)), 3)
print(Mx2, "- Мат. ожидание по x")

My2 = round((1/len(y2)*sum(y2)), 3)
print(My2, "- Мат. ожидание по y")
```

#### Оценки начальных моментов второго порядка по каждой переменной

```python
sqr_x2 = [round(x**2, 3) for x in x2]
sqr_y2 = [round(y**2, 3) for y in y2]

# Начальный момент второго порядка: a2(x) =1/n *(sum((Xi)^2))
a2_x = round((1/len(x2)*sum(sqr_x2)), 3)
print(a2_x, "- начальный момент второго порядка по х")

a2_y = round((1/len(y2)*sum(sqr_y2)), 3)
print(a2_y, "- начальный момент второго порядка по у")
```

#### Оценка смешанного начального момента второго порядка

```python
x2_multiply_y2 = [round(x*y, 3) for x,y in zip(x2,y2)]

# Cмешанный начальный момент второго порядка: a2(x,y) =1/n *(sum((Xi*Yi)))
a2_xy = round((1/len(x2)*sum(x2_multiply_y2)), 3)
print(a2_xy, "- смешанный начальный момент второго порядка")
```

#### Оценка дисперсий

```python
# Дисперсия Dx = n/(n-1) * a2(x) - n/(n-1) * (Mx)^2
Dx2 = round(len(x2)/(len(x2) -1) * a2_x - len(x2)/(len(x2) -1) * (Mx2**2), 3)
print(Dx2, "- Дисперсия по x")
Dy2 = round(len(y2)/(len(y2) -1) * a2_y - len(y2)/(len(y2) -1) * (My2**2), 3)
print(Dy2, "- Дисперсия по y")
```

#### Оценка корреляционного момента

```python
# Корреляционный момент
Kxy2 = round(len(y2)/(len(y2) -1) * a2_xy - len(y2)/(len(y2) -1) * Mx2 * My2, 3)
print(Kxy2, "- Корреляционный момент")
```
